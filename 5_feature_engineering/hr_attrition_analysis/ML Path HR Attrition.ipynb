```python
x = {"Yes":1, "No":0}
df['Attrition'].replace(x)
```

```python
object_list = df.select_dtypes(include='object').columns.tolist()
```

```python
df_encoded = pd.get_dummies(df, columns=object_list, drop_first=True)
```

```python
df_encoded.columns
```

```python
df_encoded['Attrition_Yes']
```

```python
df_encoded['Attrition_Yes'] = df_encoded['Attrition_Yes'].astype(int)
```

```python
from sklearn.model_selection import train_test_split

X = df_encoded.drop('Attrition_Yes', axis=1)
y = df_encoded['Attrition_Yes']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
numeric_cols = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()

numeric_cols.remove('EmployeeNumber') 
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test[numeric_cols] = scaler.transform(X_test[numeric_cols])
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(class_weight = 'balanced', max_iter=1000, random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=100, 
    random_state=42,
    class_weight='balanced'  # To help with imbalance
)

rf_model.fit(X_train, y_train)
```

```python
y_pred_rf = rf_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
print("\nClassification Report:\n", classification_report(y_test, y_pred_rf))
```
